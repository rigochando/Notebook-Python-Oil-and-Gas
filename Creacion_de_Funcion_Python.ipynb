```python
def z_papay(Ppr, Tpr):
  z = 1-(3.53*Ppr)/(10**(0.9813*Tpr))+(0.274*Ppr**2)/(10**(0.8157*Tpr))

  return(z)

z = z_papay(1,1)
print(z)

Ppr = list(range(0, 8, 1))
Tpr = 1.6
Tpr2 = 1.3

zcal = [z_papay(element,Tpr) for element in Ppr]
zcal2 = [z_papay(element,Tpr2) for element in Ppr]

print(zcal)
print(zcal2)

```
