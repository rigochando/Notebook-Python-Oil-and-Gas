```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_excel("Volve production data.xlsx")
df.head() 
```

```python
df["WELL_BORE_CODE"].unique()
```

```python

F_14 = df[df["WELL_BORE_CODE"] == 'NO 15/9-F-14 H']
```

```python
plt.figure(figsize= (10,5))

plt.plot( F_14["DATEPRD"], F_14["BORE_OIL_VOL"],label="Aceite",color="green") 
plt.plot(F_14["DATEPRD"],F_14["BORE_WAT_VOL"],label="Agua",color="blue");

# setting axes (x and y) labels 
plt.xlabel("Fecha", fontsize= 13, labelpad = 20)
plt.ylabel("Gasto de aceite y agua", fontsize= 13, labelpad = 20,loc="center")

# setting figure title 
plt.title("Producción de aceite y gas", fontsize= 20, pad = 20,loc="left")

plt.legend();
```

```python
# Grafico de gasto de aceite y agua

wells = df["NPD_WELL_BORE_NAME"].unique()

plt.figure(figsize=(12,20))

for i, well in enumerate(wells):
    d = df[df["NPD_WELL_BORE_NAME"]==well]
    plt.subplot(len(wells),1,i+1)
    plt.plot(d["DATEPRD"],d["BORE_OIL_VOL"],label="Aceite",color="green")
    plt.plot(d["DATEPRD"],d["BORE_WAT_VOL"],label="Agua",color="blue")
    plt.title(well)
    plt.xlabel("Fecha")
    plt.ylabel("Gasto de aceite y agua (bpd)")
    plt.legend()

#Ajuste el marco entre y alrededor de las subgraficas
plt.tight_layout()
plt.show
```
